## By Reservation Date

### By Month

#### SCM

```r
by_month = read.csv('~/repos/airbnb/long_reservation_date_by_month.csv')
nights_by_month_nr = augsynth(nights ~ trt, unit, month, by_month, progfunc="None", scm = T)
print(summary(nights_by_month_nr))
print(nights_by_month_nr$weights)
plot(nights_by_month_nr)
```

```r
by_month$log_nights = log(by_month$nights)
log_nights_by_month_nr = augsynth(log_nights ~ trt, unit, month, by_month, progfunc="None", scm = T)
print(summary(log_nights_by_month_nr))
print(log_nights_by_month_nr$weights)
plot(log_nights_by_month_nr)
```

```r
by_month$log_price = log(by_month$price)
log_price_by_month_nr = augsynth(log_price ~ trt, unit, month, by_month, progfunc="None", scm = T, fixedeff = T)
print(summary(log_price_by_month_nr))
print(log_price_by_month_nr$weights)
plot(log_price_by_month_nr)
```

#### Ridge

```r
nights_by_month_r = augsynth(nights ~ trt, unit, month, by_month, progfunc="Ridge", scm = F)
print(nights_by_month_r$lambdas)
plot(nights_by_month_r, cv = T)
nights_by_month_r = augsynth(nights ~ trt, unit, month, by_month, progfunc="Ridge", scm = F, lambda = 1e+6)
print(summary(nights_by_month_r))
print(nights_by_month_r$weights)
plot(nights_by_month_r)
```

```r
log_nights_by_month_r = augsynth(log_nights ~ trt, unit, month, by_month, progfunc="Ridge", scm = F)
print(log_nights_by_month_r$lambdas)
plot(log_nights_by_month_r, cv = T)
log_nights_by_month_r = augsynth(log_nights ~ trt, unit, month, by_month, progfunc="Ridge", scm = F, lambda = 0.1)
print(summary(log_nights_by_month_r))
print(log_nights_by_month_r$weights)
plot(log_nights_by_month_r)
```

```r
log_price_by_month_r = augsynth(log_price ~ trt, unit, month, by_month, progfunc="Ridge", scm = F)
print(log_price_by_month_r$lambdas)
plot(log_price_by_month_r, cv = T)
log_price_by_month_r = augsynth(log_price ~ trt, unit, month, by_month, progfunc="Ridge", scm = F, lambda = 0.1)
print(summary(log_price_by_month_r))
print(log_price_by_month_r$weights)
plot(log_price_by_month_r)
```
